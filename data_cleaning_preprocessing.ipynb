### Remove rows without target variables

```python
df_train = df_train[df_train['disorder_subclass'].isnull()!=True]
```

```python
disorder_mapping = {
        "Leber's hereditary optic neuropathy": "Mitochondrial genetic inheritance disorders",
        "Leigh syndrome": "Mitochondrial genetic inheritance disorders",
        "Mitochondrial myopathy": "Mitochondrial genetic inheritance disorders",
        "Alzheimer's": "Multifactorial genetic inheritance disorders",
        "Cancer": "Multifactorial genetic inheritance disorders",
        "Diabetes": "Multifactorial genetic inheritance disorders",
        "Cystic fibrosis": "Single-gene inheritance diseases",
        "Hemochromatosis": "Single-gene inheritance diseases",
        "Tay-Sachs": "Single-gene inheritance diseases",
    }
```

```python
def fill_missing_genetic_disorder(row):
    if pd.isnull(row["genetic_disorder"]) and row["disorder_subclass"] in disorder_mapping:
        return disorder_mapping[row["disorder_subclass"]]
    return row["genetic_disorder"]

df_train["genetic_disorder"] = df_train.apply(fill_missing_genetic_disorder, axis=1)
```

```python
df_train = df_train.dropna(subset=["genetic_disorder", "disorder_subclass"])
```

### Filling Null Values

```python
grouped = df_train.groupby(["genetic_disorder", "disorder_subclass"])

def fill_nulls(group):
    for column in group.columns:
        if group[column].isnull().any():
            if group[column].dtype == 'object':  
                mode_value = group[column].mode()
                if not mode_value.empty:
                    group[column] = group[column].fillna(mode_value[0])
            else: 
                median_value = group[column].median()
                group[column] = group[column].fillna(median_value)
    return group

df_train = grouped.apply(fill_nulls).reset_index(drop=True)
```

### Replacing Ambiguous Values

```python
missing_values = ["No record", "Not available", "Not applicable", "-", "Ambiguous"]

categorical_cols_train = df_train.select_dtypes(include='object').columns
categorical_cols_test = df_test.select_dtypes(include='object').columns

for col in categorical_cols_train:
    df_train[col] = df_train[col].replace(missing_values, 'Missing')

for col in categorical_cols_test:
    df_test[col] = df_test[col].replace(missing_values, 'Missing')
```

```python
df_train['symptom_total'] = df_train[['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5']].sum(axis=1).astype(int)
df_test['symptom_total'] = df_test[['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5']].sum(axis=1).astype(int)

df_train.drop(columns=['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5'], inplace=True)
df_test.drop(columns=['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5'], inplace=True)
```

```python
df_train['total_blood_cells'] = df_train['RBC_count'] + df_train['white_blood_cell_count']
df_train.drop(columns=['RBC_count', 'white_blood_cell_count'], inplace=True)

df_test['total_blood_cells'] = df_test['RBC_count'] + df_test['white_blood_cell_count']
df_test.drop(columns=['RBC_count', 'white_blood_cell_count'], inplace=True)
```

### Combined Disorder Column

```python
df_train['disorder_type'] = df_train.apply(
    lambda row: f"{row['genetic_disorder'].replace(' ', '_')}_{row['disorder_subclass'].replace(' ', '_')}", axis=1)
```
